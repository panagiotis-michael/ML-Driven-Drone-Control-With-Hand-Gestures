## <ins>Data Import And Preprocessing</ins>

### Getting Datasets From Directory

```python
dataset_training = image_dataset_from_directory( # Dataset for training
    'training_data\\classes', # Classes directory
    labels='inferred', # Replaced with numbers starting from 0
    label_mode="int", # Encoded as integer
    color_mode="grayscale", # Reduce color depth to 1
    image_size=(64, 64), # Image dimensions
    batch_size=32, 
    shuffle=True, # Shuffle the images
    seed=132,
    validation_split=0.2, # Only used if dataset is for validation purposes
    subset='training', # Training dataset
)

dataset_validation = image_dataset_from_directory(# Dataset for training validation
    'training_data\\classes',
    labels='inferred',
    label_mode="int", 
    color_mode="grayscale",
    image_size=(64, 64), 
    batch_size=32,
    shuffle=True,
    seed=132,
    validation_split=0.2, # Used for validation
    subset='validation', # Validation dataset
)

dataset_testing = image_dataset_from_directory(# Dataset for testing
    'testing_data\\classes',
    labels='inferred', 
    label_mode="int",
    color_mode="grayscale",
    image_size=(64, 64),
    batch_size=32,
    shuffle=False # Shuffle must not be True
)
```

## <ins>CNN Model</ins>

### Model Structure

```python
inputs = Input(shape=(64,64,1))
rescale = Rescaling(1./255)(inputs)
conv1 = Conv2D(filters=32, kernel_size=(3,3))(rescale)
pool1 = MaxPooling2D(pool_size=(2,2))(conv1)
activation1 = Activation('relu')(pool1) # Activation function after maxpooling for better performance
norm1 = BatchNormalization()(activation1) # Better optimization
conv2 = Conv2D(filters=128, kernel_size=(3,3))(norm1)
pool2 = MaxPooling2D(pool_size=(3,3))(conv2)
activation2 = Activation('relu')(pool2)
norm2 = BatchNormalization()(activation2)
conv3 = Conv2D(filters=64, kernel_size=(2,2))(norm2)
pool3 = MaxPooling2D(pool_size=(2,2))(conv3)
activation3 = Activation('relu')(pool3)
norm3 = BatchNormalization()(pool3)
flat1 = Flatten()(norm3)
dense = Dense(units=32)(flat1)
activation4 = Activation('relu')(dense)
outputs = Dense(units=5, activation='softmax')(activation4)
model = Model(inputs=inputs, outputs=outputs)
```

### Model Summary

```python
model.summary()
```

### Model Compilation

```python
model.compile(
  optimizer='adam',
  loss='sparse_categorical_crossentropy',
  metrics=['accuracy'])
```

### Learning Rate Adjustment

```python
learning_rate_adjustment = ReduceLROnPlateau(monitor='val_accuracy', patience = 2, verbose=1,factor=0.5, min_lr=0.00001) # -> lower bound
```

### Fitting the Model

```python
model_history=model.fit(
  dataset_training,
  validation_data=dataset_validation,
  callbacks=learning_rate_adjustment,
  epochs=5
)
```

### Model Accuracy Plot

```python
plt.figure(figsize=(10, 8))
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['Training', 'Validation'])
plt.show()
```

### Model Loss Plot

```python
plt.figure(figsize=(10, 8))
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['Training', 'Validation'])
plt.show()
```

### Accuracy, Precision, Recall, F1-Score Heatmap

```python
y_test = np.array([])

for image, label in dataset_testing: # Get true labels
    y_test=np.concatenate([y_test,label])

y_pred = model.predict(dataset_testing,verbose=1) # Get predictions
y_pred = y_pred.argmax(axis=-1)

target_names=['open_palm','pointing_down','pointing_left','pointing_right','pointing_up']
report = classification_report(y_test, y_pred, target_names=target_names, output_dict=True) # Get the report
plt.figure(figsize=(10, 8))
sns.heatmap(pd.DataFrame(report).T, annot=True, cmap='Blues') # Setup heatmap with the help of pandas
```

### Confusion Matrix

```python
conf_matrix = confusion_matrix(y_test, y_pred)
index = ['open_palm','pointing_down','pointing_left','pointing_right','pointing_up']  
columns = ['open_palm','pointing_down','pointing_left','pointing_right','pointing_up']
conf_matrix_df = pd.DataFrame(conf_matrix,columns,index)
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(conf_matrix_df, annot=True, cmap='Blues',ax=ax)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

## <ins>Drone Section</ins>

### Image Classification

```python
def predict_image(model,img,img_width,img_height):
    img = cv2.resize(img, (img_width, img_height)) # Resizing image to desired dimensions (64,64)
    img = tf.image.rgb_to_grayscale(img) # Transforming image to grayscale to match model demands
    img_array = image.img_to_array(img) # Turning image to array form
    img_batch = np.expand_dims(img_array, axis=0) # Using numpy create batch from array
    prediction = model.predict(img_batch).argmax() # Get prediction
    return prediction
```

### Predict Movement

```python
def predict_movement(prediction):
    if prediction == 0: # Hold still option
        return
    elif prediction == 1:
        drone.move_down(30) # Parameters here are the amount of centimeters to move.
    elif prediction == 2:
        drone.move_left(30)
    elif prediction == 3:
        drone.move_right(30)
    elif prediction == 4:
        drone.move_up(30)
```

### Initializing Drone Object

```python
drone = tello.Tello()
```

### Connecting To The Drone

```python
drone.connect()
print(drone.get_battery())
```

### Load Model

```python
final_model = tf.keras.models.load_model('model')
```

### Test With Computer Camera

```python
cap = cv2.VideoCapture(0) # Opening camera

drone.takeoff() # Take off drone

while True:
    ret, frame = cap.read() # Take frame from camera
    cv2.imshow("Frame", frame) # Show on separate window
    prediction = predict_image(final_model,frame,64,64) # Get prediction
    predict_movement(prediction) # Move according to prediction
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Actions if 'q' is pressed
        cap.release() # Turn off camera
        drone.land() # Land the drone
        break
        
cv2.destroyAllWindows() # Close frame display window
```
